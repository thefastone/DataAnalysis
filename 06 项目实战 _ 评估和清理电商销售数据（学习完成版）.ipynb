# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

####这里发现单价为0，那么是不是代表了无效数据？——不一定，因为中间有省略，所以需要考虑在此之上有没有单价不为0的数据呢？操作的时候就是合并两个条件

```python
##也就是说所有单价为0的商品商品名也是空的，针对需要分析热销商品的目标来看是无意义的，所以可以舍弃
```

#####客户信息并不会影响我们对于商品热销程度的分析，所以可以保留这些缺失值

####评估重复数据——变量重复不等于有问题（比如日期）

```python
df1["InvoiceNo"].duplicated()
```

这个地方要注意，发票标识虽然唯一，但是一个发票可以包含多个商品，所以发票标识在表格 中并不会只出现一次，同时一个商品也可被多次购买，一个客户也可以多次购买商品，因此对于这个数据中，并没有唯一值的要求

###评估不一致数据

```python
df1["Country"].value_counts()
```

####发现问题了，这里英国有United Kingdom、UK以及U.K.三种，美国则是有USA、United States两种，所以需要修改

```python
df1["Country"].replace(["UK","U.K."],"United Kingdom",inplace=True)
df1["Country"].replace("United States","USA",inplace=True)
```

```python
df1["Country"].value_counts()
```

####还需要评估无效以及错误数据——使用`describe`方法

```python
df1.describe()
```

```python
这里可以看到数量以及单价的最小值存在负数，这是不应该出现的
```

#### 先将数量和单价的负值筛选出来观察一下

```python
df1[df1["Quantity"]<0]
```

**首先观察到这些发票抬头都为C，代表了退货，那是否可以认定所有开头为C的数量都为负数呢？**

```python
df1[(df1["Quantity"]<0) & (df1["InvoiceNo"].str[0] !="C")]
```

上面的猜想是错误的，但是可以发现，不以C开头数量为负数的数据，其单价都为0，由此可推断是否这样的数据都为0？

```python
df1[(df1["Quantity"]<0) & (df1["InvoiceNo"].str[0] !="C") & (df1["UnitPrice"]!=0)]
```

由此可见，所有有问题的数据其单价都为0，所以推测可以舍弃这部分数据？

再次总结，当数量为负数的时候，分两种情况，第一种就是订单被取消了（以C开头），要么就是单价为0；其实这些数据都不是有效数据，因为本次分析的目的是为了获取畅销商品

```python
df1[df1["UnitPrice"]<0]
```

### 从description来看，这个属于坏账调整，也不是商品数据，所以也是可以删除掉的

## 清理数据

## 从评估结果可以发现，Quantity为负数，以及UnitPrice为0以及为负数的数据都属于无效数据，需要进行删除

### 整理一下待办事项
- 把`InvoiceDate`变量的数据类型转换为日期时间
- 把`CustomerID`变量的数据类型转换为字符串
- 把`Description`变量缺失的观察值删除
- 替换英国和美国的重复数据（已完成）
- 把`Quantity`变量为复数的观察值删除
- 把`UnitPrice`变量为负数的观察值删除

```python
cleaned_data = df1.copy()
cleaned_data.head()
```

```python
cleaned_data["InvoiceDate"] =pd.to_datetime(cleaned_data["InvoiceDate"])
cleaned_data["InvoiceDate"]
```

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].astype(str)
cleaned_data["CustomerID"]
#转换成字符串后有个.0结尾，这个就很奇怪
```

```python
cleaned_data["CustomerID"] =cleaned_data["CustomerID"].str.slice(0,-2)
```

```python
cleaned_data = cleaned_data.dropna(subset=["Description"])
```

```python
cleaned_data["Description"].isnull().sum()
```

```python
cleaned_data["Country"] = cleaned_data["Country"].replace("USA","United States")
cleaned_data["Country"].value_counts()
```

### 目前还没学按照条件删除，但是可以筛选出quantity大于0 的情况然后赋值给原来的DataFrame就可以了

```python
df2 = cleaned_data[cleaned_data["Quantity"]>0]
len(df2[df2["Quantity"]<=0])
cleaned_data = df2
```

```python
cleaned_data = cleaned_data[cleaned_data["UnitPrice"]>0]
len(cleaned_data[cleaned_data["UnitPrice"]<=0])
cleaned_data
```
